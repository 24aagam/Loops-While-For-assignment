```python
# 1. Print number from 1 to 5 using a while loop.

num = 1

while num <= 5:
  print(num)
  num += 1
```

```python
# 2. Calculate the sum of numbers from 1 to 10 using a while loop.

num = 1

sum = 0

while num <= 10:
  sum += num

  num += 1

print(sum)
```

```python
# 3. Calculate the factorial of a number using a for loop.

n = 4

factorial = 1

for i in range(1, n+1):
  factorial *= i


print(factorial)
```

```python
# 4. Count the number of vowels in a string using a for loop.

vowel = 'a','e','i','o','u'

for i in range(5):
  count = i+1, vowel[i]
  print(count)
```

```python
# 5. Print a pattern using nested loops.

for i in range(5):
  for j in range(1, i+1):
    print(j, end=' ')
  print()
```

```python
rows = 1
columns = 1
while rows <= 5:
  while columns <= rows:
    print("*"*columns, end=' ')
    columns += 1
  print()
  rows += 1
```

```python
# 6. Generate a multiplication table using nested loops.

num = 6

for i in range(1, num):
  for j in range(1, 11):
    multiplication = i * j
    print(f"{i} * {j} = {multiplication}")
  print()
```

```python
row = 1

while row < 6:
  col = 1
  while col < 11:
    print(f"{row} * {col} = {row * col}")
    col +=1
  print()
  row += 1
```
